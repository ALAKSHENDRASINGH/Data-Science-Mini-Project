# Backward elimination with P- values

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 1: Setup Your Data (Using a sample DataFrame) ---
# In your code, you would load your actual DataFrame. This is for demonstration.
crime_data = {
    'air_water_prevention_pollution_control': np.random.randint(0, 100, 100),
    'liquor_narcotic_drugs_prohibition': np.random.randint(0, 100, 100),
    'gambling_act': np.random.randint(0, 100, 100),
    'essential_commodities': np.random.randint(0, 100, 100),
    'medical_termination_of_pregnancy': np.random.randint(0, 100, 100),
    'prevention_of_atrocities_against_sts': np.random.randint(0, 100, 100),
    'registration_of_foreigners_and_foreigners': np.random.randint(0, 100, 100),
    'citizenship_act': np.random.randint(0, 100, 100),
    'indecent_representation_of_women_prohibition': np.random.randint(0, 100, 100),
    'wildlife_protection': np.random.randint(0, 100, 100),
    'child_labour_prohibition_regulation': np.random.randint(0, 100, 100),
    'women_protection_from_domestic_voilence': np.random.randint(0, 100, 100),
    'other_state_local_acts': np.random.randint(0, 100, 100),
    'other_crime_1': np.random.randint(0, 500, 100)
}
df = pd.DataFrame({
    'State': ['State']*100, 'District': ['District']*100, 'Year': [2025]*100, **crime_data
})

# Create the 'total_crimes' column by summing ALL available crime columns
crime_columns = df.columns[3:]
df['total_crimes'] = df[crime_columns].sum(axis=1)

# --- Step 2: Start the Backward Elimination Process ---
initial_features = [
    'air_water_prevention_pollution_control', 'liquor_narcotic_drugs_prohibition',
    'gambling_act', 'essential_commodities', 'medical_termination_of_pregnancy',
    'prevention_of_atrocities_against_sts', 'registration_of_foreigners_and_foreigners',
    'citizenship_act', 'indecent_representation_of_women_prohibition',
    'wildlife_protection', 'child_labour_prohibition_regulation',
    'women_protection_from_domestic_voilence', 'other_state_local_acts'
]
y = df['total_crimes']
features_to_keep = initial_features.copy()
p_threshold = 0.05

print("--- Starting Backward Elimination ---")

while len(features_to_keep) > 0:
    X = df[features_to_keep]
    X_with_const = sm.add_constant(X)
    model = sm.OLS(y, X_with_const).fit()
    
    p_values = model.pvalues.drop('const')
    max_p_value = p_values.max()
    
    if max_p_value > p_threshold:
        feature_to_remove = p_values.idxmax()
        print(f"Removing feature: {feature_to_remove} (P-value: {max_p_value:.4f})")
        
        # --- GRAPH GENERATION (During Removal) ---
        plt.figure(figsize=(8, 8))
        sns.scatterplot(x=model.fittedvalues, y=y, alpha=0.6)
        plt.xlabel('Predicted Total Crimes')
        plt.ylabel('Actual Total Crimes')
        plt.title(f'Model Performance Before Removing "{feature_to_remove}"\nR-squared: {model.rsquared:.3f}')
        max_val = max(y.max(), model.fittedvalues.max())
        min_val = min(y.min(), model.fittedvalues.min())
        plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--')
        plt.grid(True)
        plt.show()
        plt.close()
        
        features_to_keep.remove(feature_to_remove)
    else:
        print("\nAll remaining features are statistically significant.")
        break

# --- Step 3: Display the Final, Optimized Model and its Graphs ---
print("\n--- Final Model Summary (After Backward Elimination) ---")
final_model = sm.OLS(y, sm.add_constant(df[features_to_keep])).fit()
print(final_model.summary())

# --- FINAL GRAPH 1: Partial Regression Plots for the Final Model ---
print("\nGenerating Partial Regression Plots for the final significant features...")
fig = plt.figure(figsize=(12, 12))
sm.graphics.plot_partregress_grid(final_model, fig=fig)
plt.suptitle('Partial Regression Plots for Final Model', fontsize=20, y=1.02)
plt.show()
plt.close()

# --- FINAL GRAPH 2: Actual vs. Predicted for the Final Model ---
print("Generating final Actual vs. Predicted plot...")
plt.figure(figsize=(8, 8))
sns.scatterplot(x=final_model.fittedvalues, y=y, alpha=0.6)
plt.xlabel('Predicted Total Crimes (Final Model)')
plt.ylabel('Actual Total Crimes')
plt.title(f'Final Model Performance\nAdj. R-squared: {final_model.rsquared_adj:.3f}')
max_val = max(y.max(), final_model.fittedvalues.max())
min_val = min(y.min(), final_model.fittedvalues.min())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--')
plt.grid(True)
plt.show()
plt.close()
```

# Backward elimination with Adjusted R squared values

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 1: Setup Your Data (Using a sample DataFrame) ---
# In your code, you would load your actual DataFrame.
crime_data = {
    'air_water_prevention_pollution_control': np.random.randint(0, 100, 100),
    'liquor_narcotic_drugs_prohibition': np.random.randint(0, 100, 100),
    'gambling_act': np.random.randint(0, 100, 100),
    'essential_commodities': np.random.randint(0, 100, 100),
    'medical_termination_of_pregnancy': np.random.randint(0, 100, 100),
    'prevention_of_atrocities_against_sts': np.random.randint(0, 100, 100),
    'registration_of_foreigners_and_foreigners': np.random.randint(0, 100, 100),
    'citizenship_act': np.random.randint(0, 100, 100),
    'indecent_representation_of_women_prohibition': np.random.randint(0, 100, 100),
    'wildlife_protection': np.random.randint(0, 100, 100),
    'child_labour_prohibition_regulation': np.random.randint(0, 100, 100),
    'women_protection_from_domestic_voilence': np.random.randint(0, 100, 100),
    'other_state_local_acts': np.random.randint(0, 100, 100),
    'other_crime_1': np.random.randint(0, 500, 100)
}
df = pd.DataFrame({
    'State': ['State']*100, 'District': ['District']*100, 'Year': [2025]*100, **crime_data
})

# Create the 'total_crimes' column by summing ALL available crime columns
crime_columns = df.columns[3:]
df['total_crimes'] = df[crime_columns].sum(axis=1)

# --- Step 2: Start the Backward Elimination Process using Adjusted R-squared ---
initial_features = [
    'air_water_prevention_pollution_control', 'liquor_narcotic_drugs_prohibition',
    'gambling_act', 'essential_commodities', 'medical_termination_of_pregnancy',
    'prevention_of_atrocities_against_sts', 'registration_of_foreigners_and_foreigners',
    'citizenship_act', 'indecent_representation_of_women_prohibition',
    'wildlife_protection', 'child_labour_prohibition_regulation',
    'women_protection_from_domestic_voilence', 'other_state_local_acts'
]
y = df['total_crimes']
features_to_keep = initial_features.copy()

initial_model = sm.OLS(y, sm.add_constant(df[features_to_keep])).fit()
best_adj_rsquared = initial_model.rsquared_adj

print(f"--- Starting Backward Selection based on Adjusted R-squared ---")
print(f"Initial Model Adj. R-squared: {best_adj_rsquared:.4f} with {len(features_to_keep)} features\n")

while True:
    candidate_models = []
    
    for feature_to_test in features_to_keep:
        temp_features = features_to_keep.copy()
        temp_features.remove(feature_to_test)
        
        if not temp_features: # Avoids error if only one feature is left
            continue
            
        temp_model = sm.OLS(y, sm.add_constant(df[temp_features])).fit()
        candidate_models.append({'removed_feature': feature_to_test, 'adj_rsquared': temp_model.rsquared_adj})

    if not candidate_models:
        break

    best_candidate = max(candidate_models, key=lambda x: x['adj_rsquared'])
    
    if best_candidate['adj_rsquared'] > best_adj_rsquared:
        feature_to_remove = best_candidate['removed_feature']
        best_adj_rsquared = best_candidate['adj_rsquared']
        features_to_keep.remove(feature_to_remove)
        print(f"Removing '{feature_to_remove}', New Best Adj. R-squared: {best_adj_rsquared:.4f}")
    else:
        print("\nNo further improvement in Adj. R-squared possible. Stopping.")
        break

# --- Step 3: Display the Final, Optimized Model and its Graphs ---
print("\n--- Final Model Summary (Based on Adjusted R-squared) ---")
final_model = sm.OLS(y, sm.add_constant(df[features_to_keep])).fit()
print(final_model.summary())

# --- Step 4: Generate and Display Graphs for the Final Model ---

# FINAL GRAPH 1: Partial Regression Plots
print("\nGenerating Partial Regression Plots for the final significant features...")
fig = plt.figure(figsize=(12, 12))
sm.graphics.plot_partregress_grid(final_model, fig=fig)
plt.suptitle('Partial Regression Plots for Final Model', fontsize=20, y=1.02)
plt.show()
plt.close()

# FINAL GRAPH 2: Actual vs. Predicted Plot
print("Generating final Actual vs. Predicted plot...")
plt.figure(figsize=(8, 8))
sns.scatterplot(x=final_model.fittedvalues, y=y, alpha=0.6)
plt.xlabel('Predicted Total Crimes (Final Model)')
plt.ylabel('Actual Total Crimes')
plt.title(f'Final Model Performance\nAdj. R-squared: {final_model.rsquared_adj:.3f}')
max_val = max(y.max(), final_model.fittedvalues.max())
min_val = min(y.min(), final_model.fittedvalues.min())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--')
plt.grid(True)
plt.show()
plt.close()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 1: Setup Your Data (Using a sample DataFrame) ---
# In your code, you would load your actual DataFrame. This is for demonstration.
crime_data = {
    'liquor_narcotic_drugs_prohibition': np.random.randint(0, 100, 100),
    'medical_termination_of_pregnancy': np.random.randint(0, 100, 100),
    'registration_of_foreigners_and_foreigners': np.random.randint(0, 100, 100),
    'citizenship_act': np.random.randint(0, 100, 100),
    'indecent_representation_of_women_prohibition': np.random.randint(0, 100, 100),
    'wildlife_protection': np.random.randint(0, 100, 100),
    'child_labour_prohibition_regulation': np.random.randint(0, 100, 100),
    'other_state_local_acts': np.random.randint(0, 100, 100),
    # Add other crimes to make the total more realistic
    'other_crime_1': np.random.randint(0, 500, 100)
}
df = pd.DataFrame({
    'State': ['State']*100, 'District': ['District']*100, 'Year': [2025]*100, **crime_data
})

# Create the 'total_crimes' column by summing ALL available crime columns
crime_columns = df.columns[3:]
df['total_crimes'] = df[crime_columns].sum(axis=1)

# --- Step 2: Define Variables for the MLR Model ---
# This list contains the features from your final model (optimized for Adj. R-squared).
features_for_model = [
    'liquor_narcotic_drugs_prohibition',
    'medical_termination_of_pregnancy',
    'registration_of_foreigners_and_foreigners',
    'citizenship_act',
    'indecent_representation_of_women_prohibition',
    'wildlife_protection',
    'child_labour_prohibition_regulation',
    'other_state_local_acts'
]

# Define the dependent (y) and independent (X) variables
y = df['total_crimes']
X = df[features_for_model]

# Add a constant (intercept) to the model
X_with_const = sm.add_constant(X)

# --- Step 3: Fit the MLR Model and Print the Summary ---
final_model = sm.OLS(y, X_with_const).fit()

print("--- MLR Model Summary (Based on Adj. R-squared Selection) ---")
print(final_model.summary())

# --- Step 4: Generate and Display Graphs for the Model ---

# Graph 1: Partial Regression Plots
print("\nGenerating Partial Regression Plots...")
fig = plt.figure(figsize=(12, 10))
sm.graphics.plot_partregress_grid(final_model, fig=fig)
plt.suptitle('Partial Regression Plots for Each Feature', fontsize=16, y=1.02)
plt.show()
plt.close()

# Graph 2: Actual vs. Predicted Plot
print("Generating Actual vs. Predicted plot...")
plt.figure(figsize=(8, 8))
sns.scatterplot(x=final_model.fittedvalues, y=y, alpha=0.6)
plt.xlabel('Predicted Total Crimes (Fitted Values)')
plt.ylabel('Actual Total Crimes')
plt.title(f'Model Performance\nAdj. R-squared: {final_model.rsquared_adj:.3f}')
max_val = max(y.max(), final_model.fittedvalues.max())
min_val = min(y.min(), final_model.fittedvalues.min())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--')
plt.grid(True)
plt.show()
plt.close()
```
